```python
def normalizar_texto(texto):
    if pd.isnull(texto):
        return texto
    texto = str(texto).lower()
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    return texto.strip()
```

```python
df['produto'] = df['produto'].apply(normalizar_texto)
df['marca'] = df['marca'].apply(normalizar_texto)
```

```python
def normalizar_produto(texto):
    if pd.isnull(texto):
        return texto
    texto = str(texto).lower()
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    texto = re.sub(r'[^a-z0-9\s]', '', texto)
    return texto.strip()
```

```python
df['produto'] = df['produto'].apply(normalizar_produto)
```

```python
correcoes = {
    'refrigkrante': 'refrigerante',
    'scl': 'sal',
    'agua mineras': 'agua mineral',
    'presuntd': 'presunto'
}

df['produto'] = df['produto'].replace(correcoes)
```

```python
def normalizar_marca(texto):
    if pd.isnull(texto):
        return texto
    texto = str(texto).lower()
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    texto = texto.replace('-', ' ')
    texto = re.sub(r'[^a-z0-9\s]', '', texto)
    return texto.strip()

df['marca'] = df['marca'].apply(normalizar_marca)
```

```python
correcoes_marca = {
    'guarana antartica': 'guarana antarctica',
    'ki brasa': 'kibrasa'
}

df['marca'] = df['marca'].replace(correcoes_marca)
```

```python
def validar_cep(cep):
    if pd.isnull(cep):
        return False
    cep = str(cep).strip()
    return bool(re.match(r'^\d{5}-?\d{3}$', cep))
```

```python
df['cep_valido'] = df['cep'].apply(validar_cep)
```
